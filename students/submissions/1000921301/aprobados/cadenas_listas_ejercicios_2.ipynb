```python
from urllib.request import urlopen
response = urlopen('https://www.gutenberg.org/files/74/74-0.txt')
book = str(response.read())
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cap = book.split("CHAPTER")
jim = np.char.count(cap, "Jim")
tom = np.char.count(cap, "Tom")
huck = np.char.count(cap, "Huck")
chapters = np.linspace(1, 35, 35)
```

```python
fig, ax= plt.subplots(1, 1)

plt.style.use("seaborn")
ax.plot(chapters, jim[36:], color = "darkgreen", label = "$Jim$")
ax.plot(chapters, tom[36:], color = "darkred", label = "$Tom$")
ax.plot(chapters, huck[36:], color = "darkblue", label = "$Huck$")
plt.xlabel("Captulos")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de nombres por captulo")

ax.legend()
```
