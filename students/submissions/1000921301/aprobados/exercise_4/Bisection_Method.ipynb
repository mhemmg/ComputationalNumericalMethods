```python
import numpy as np
import matplotlib.pyplot as plt

def parabola(x):
  return x**2 - 11

def bisection(xl, xr, tol, f): # xl: Valor a la izquierda, xr: Valor a la derecha
  L = []                       # tol: Tolerancia, f: Función.
  val = abs(f((xl+xr)/2))
  while val > tol:
    xn = (xl+xr)/2
    L.append(xn)
    if f(xl) * f(xn) < 0:
      xr = xn
    else:
      xl = xn
    val = abs(f((xl+xr)/2))
  return L

yl = bisection(3, 4, 0.0001, parabola)
xl = np.linspace(0, len(yl), len(yl))

plt.style.use("seaborn")
fig, ax =  plt.subplots()

ax.plot(xl, yl, label = f"$Valor$ $final = {yl[-1]:.3f}$", color = "darkred")
plt.title("$Bisection$ $Method$ $(F(x) = x^2 - 11, tol = 0.0001)$")
ax.set_xlabel("$Iteración$")
ax.set_ylabel("$Valor$ $de$ $la$ $raíz$")
ax.legend()

print(f"Número de iteraciones: {len(yl)}")
```
