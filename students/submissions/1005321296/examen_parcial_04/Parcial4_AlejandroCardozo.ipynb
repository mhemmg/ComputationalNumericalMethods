````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.integrate import solve_ivp
````

## 4

````python
def lotkavolterra(t, z, a, b, c, d):
    x, y = z
    return [a*x - b*x*y, -c*y + d*x*y]
````

````python
sol = solve_ivp(lotkavolterra, [0, 30], [4, 2], args=(1, 1, 1, 1),
                dense_output=True)
````

````python
t = np.linspace(0, 30, 300)
z = sol.sol(t)
import matplotlib.pyplot as plt
plt.plot(t, z.T)
plt.xlabel('Tiempo (meses)')
plt.ylabel('Poblacio√≥n')
plt.grid()
plt.legend(['Conejos', 'Zorros'], shadow=True)
plt.title('Lotka-Volterra')
plt.show()
````

````python
x, y = z
plt.plot(x,y,'g')
plt.grid()
plt.xlabel('N√∫mero de conejos')
plt.ylabel('Numero de zorros')
````

En el grafico 1, observe como la poblaci√≥n de una especia decrece al mismo tiempo que la otra aumenta, esto se puede explicar porque posiblemete la alimentaci√≥n de los zorros depende enteramente de los conejos, por eso cuando se acabaron los conejos, los zorros empiezan a decrecer.

En el grafico 2, ilustra que esta relaci√≥n conejo-zorro es periodica, y ninguna especie debe llegar a 0 o las dos se extingiran. Observe los dos "picos" al rededor de (1,4) y (4,1), se traducen a lo que se dijo anteriormente de los picos y valles en la gr√°fica 1.
