```python
import numpy as np
def cuadratic_interp(x,y,xinterp):
    #The coefficients for the approximation:
    b_0 = y[0]
    b_1 = ((y[1]-y[0])/(x[1]-x[0]))
    b_2 = (((y[1]-y[0])/(x[2]-x[1]))- b_1)/(x[2]-x[0])
    #So the interpolation gave:
    yinterp = b_0+b_1*(x[2]-x[0])+b_2*(x[2]-x[0])*(x[2]-x[1])
    return yinterp
```
