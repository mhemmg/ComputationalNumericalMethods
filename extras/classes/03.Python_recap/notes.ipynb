# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-np.pi, np.pi, 100)
y1=np.sin(x)
y2=np.cos(x)
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(x,y1,'-b',label='sin(x)',lw=1)
ax.plot(x,y2,'.r',label='cos(x)')
ax.legend()
ax.set_title('Plot sin(x) , cos(x)')
ax.set_xlabel('x values')
ax.set_ylabel('y values')
ax.grid(True)
fig.savefig('sin_cos.jpg')
```

```python
fig, ax = plt.subplots(2,1,figsize=(6,12))

ax[0].plot(x,y1,'-b',label='sin(x)')
ax[1].plot(x,y2,'.r',label='cos(x)')

ax[0].legend()
ax[0].set_title('Plot sin(x) ')
ax[0].set_xlabel('x values')
ax[0].set_ylabel('y values')
ax[0].grid(True)

ax[1].legend()
ax[1].set_title('Plot sin(x) ')
ax[1].set_xlabel('x values')
ax[1].set_ylabel('y values')
ax[1].grid(True)
```

```python
fig, ax = plt.subplots(2,2,figsize=(6,8))

ax[0,0].plot(x,y1,'-b',label='sin(x)')
ax[1,1].plot(x,y2,'.r',label='cos(x)')
ax[0,1].plot(x,y2*y1,'.g',label='cos(x)')

ax[0,0].legend()
ax[0,0].set_title('Plot sin(x) ')
ax[0,0].set_xlabel('x values')
ax[0,0].set_ylabel('y values')
ax[0,0].grid(True)

ax[0,1].legend()
ax[0,1].set_title('Plot sin(x) ')
ax[0,1].set_xlabel('x values')
ax[0,1].set_ylabel('y values')
ax[0,1].grid(True)

ax[1,1].legend()
ax[1,1].set_title('Plot sin(x) ')
ax[1,1].set_xlabel('x values')
ax[1,1].set_ylabel('y values')
ax[1,1].grid(True)
```

# Plotting distributions

```python
mu1=0
std1=0.3
mu2=0
std2=0.2
N=1000

x=np.random.normal(mu1,std1,N)
y=np.random.normal(mu2,std2,N)

print('mean x:', np.mean(x))
print('std x:', np.std(x))

fig,ax= plt.subplots(1,1,figsize=(5,4))
ax.scatter(x,y,c='m',s=10)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
```
